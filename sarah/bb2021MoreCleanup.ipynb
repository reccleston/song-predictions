```python
import pandas as pd
```

```python
bb2021 = pd.read_csv('../data/bb21.csv')
bb2021.head()
```

```python
bb2021 = bb2021[['rank', 'artist', 'title', 'date']]
bb2021.head()
```

```python
bb2021['hitTF'] = 1
bb2021.head()
```

```python
bb2021.to_csv('../data/bb21.csv', index=False)
```

```python
bb2021['numWeeks'] = 0
bb2021['peak'] = 0
bb2021.head()
```

```python
song_counts = pd.DataFrame(bb2021.groupby(['title'])['date'].count()).reset_index()
song_counts.columns = ['title', 'numWeeks']
song_counts.head()
```

```python
top_rankings = pd.DataFrame(bb2021.groupby(['title'])['rank'].min()).reset_index()
top_rankings.columns = ['title', 'top_rank']
top_rankings.head()
```

```python
combo = pd.merge(song_counts, top_rankings, on='title')
combo.head()
```

```python
combo_dict = combo.set_index('title').T.to_dict('list')
combo_dict
```

```python
for k, v in combo_dict.items():
    print(v[0])
```

```python
bb2021.iloc[0, 1]
```

```python
numWeeksList = []
peakList = []
for x in range(0, len(bb2021)):
    for k, v in combo_dict.items():
        if bb2021.iloc[x, 2] == k:
            numWeeksList.append(v[0])
            peakList.append(v[1])
```

```python
bb2021['numWeeks'] = numWeeksList
bb2021['peak'] = peakList
bb2021.head()
```

```python
bb2021.to_csv('../data/newbb21.csv', index=False)
```
