```python
# imports
import pandas as pd
```

```python
# reading in 2021 csv
bb2021 = pd.read_csv('../data/bb21.csv')
```

```python
# renaming cols
bb2021 = bb2021[['rank', 'artist', 'title', 'date']]
bb2021.head()
```

```python
# adding in a hits t/f col, all are hits (1) here
bb2021['hitTF'] = 1
bb2021.head()
```

```python
# saving the reordered, new col csv
bb2021.to_csv('../data/bb21.csv', index=False)
```

```python
# placeholder cols for number of weeks on charts and peak rank
bb2021['numWeeks'] = 0
bb2021['peak'] = 0
bb2021.head()
```

```python
# reading in csv from previous years for getting last week of 2020 previous position vals
bb2020yon = pd.read_csv('../data/bb2020BetterCols.csv')
bb2020yon.head()
```

```python
bb2020yon['date'] =  pd.to_datetime(bb2020yon['date'])
```

```python
# noting that there are some discontinuities - 'previous position' refers to previous week's position,
#     not if the song has charted before
#     but there are also some issues with not properly noting previous position (see Perry Como blanks)
bb2020yonOrdered = pd.DataFrame(bb2020yon.sort_values(['song', 'date']))
bb2020yonOrdered.head(10)
```

```python
# getting the last month of 2020 to add on to 2021 data for checking for previous week data
bb2020shortened = bb2020yonOrdered[bb2020yonOrdered['date'].dt.date.astype(str) >= '2020-12-26'].reset_index(drop=True)
bb2020shortened.head(10)
```

```python
# renaming and reordering columns for combo w/ last month of 2020 data 
# renaming
bb2021.columns = ['chart_position', 'performer', 'song', 'date', 'hitTF', 'weeks_on_chart', 'peak']
bb2021.head()
```

```python
# creating previous position col
bb2021['previous_position'] = 0
```

```python
# reordering cols
bb2021 = bb2021[['song', 'performer','date', 'chart_position', 'previous_position', 'peak', 'weeks_on_chart', 'hitTF']]
bb2021.head()
```

```python
datafuller = pd.concat([bb2020shortened, bb2021])
datafuller.head()
```

```python
print(f'before: {datafuller["date"].dtype}')
datafuller['date'] = pd.to_datetime(datafuller['date'], utc=False)
print(f'after: {datafuller["date"].dtype}')
```

```python
# data2021 = pd.DataFrame(bb2021.sort_values(['title', 'date']))
# data2021.head(20)
datafullerSorted = pd.DataFrame(datafuller.sort_values(['song', 'date']))
datafullerSorted.head(20)
```

```python
# test_whole = []
# range starts at one because track at index 0 has no previous week to which to compare it
for x in range (1, len(datafullerSorted)):
#     if titles match...
    if datafullerSorted.iloc[x-1, 0] == datafullerSorted.iloc[x, 0]:
#         change previous position col to value for chart pos of prev week
        datafullerSorted.iloc[x, 4] = datafullerSorted.iloc[x-1, 3]
#test['previousPosition'] = test_whole
datafullerSorted.head()
```

```python
datafullerSorted.head(20)
```

```python
# now that the previous positions are filled in, remove the 2020 data
# sort by date
datafullerResort = datafullerSorted.sort_values('date')
# remove all values before 2021
dfrCut = datafullerResort[datafullerResort['date'].dt.date.astype(str) > '2020-12-31'].reset_index(drop=True)
dfrCut.head(10)
```

```python
# processing to find number of weeks each song was on the charts in 2021
# creating a df showing the song title and num of weeks through a groupby
song_counts = pd.DataFrame(dfrCut.groupby(['song'])['date'].count()).reset_index()
# renaming the cols to be descriptive
song_counts.columns = ['song', 'weeks_on_chart']
song_counts.head()
```

```python
# processing to get the top position each track has occupied
# creating a df showing song title and highest position (lowest number) each track has had using groupby
top_rankings = pd.DataFrame(dfrCut.groupby(['song'])['chart_position'].min()).reset_index()
# renaming to be more descriptive
top_rankings.columns = ['song', 'peak']
top_rankings.head()
```

```python
combo = pd.merge(song_counts, top_rankings, on='song')
combo.head()
```

```python
combo_dict = combo.set_index('song').T.to_dict('list')
combo_dict
```

```python
numWeeksList = []
peakList = []
for x in range(0, len(dfrCut)):
    for k, v in combo_dict.items():
#         if the song title matches the key value,
        if dfrCut.iloc[x, 0] == k:
            numWeeksList.append(v[0])
            peakList.append(v[1])
```

```python
dfrCut['weeks_on_chart'] = numWeeksList
dfrCut['peak'] = peakList
dfrCut.head()
```
