```python
# joining the bb dfs
# ! would have to include the spotify id and song attribute cols to these
# making the col names the same
bbTo20.columns = ['title', 'artist', 'date', 'rank', 'hitTF']
# col order
bbTo20 = bbTo20[['rank', 'artist', 'title', 'date', 'hitTF']]
bbTo20.head()
```

```python
# concatting the two bb dfs
allbbs = pd.concat([bbTo20, bb2021]).reset_index(drop=True)
allbbs.head()
```

```python
# code in progress: using currently theoretical cols in the dfs
# make sure that the spotify ID col is correctly named here
hitIDs = allbbs['spotifID'] 
updatedhitTF = []
for x in range(0, len(tracks)):
#     this is assuming that the spotifID col is the rightmost (as of right now) one in the df, ie with index 19
    if tracks.iloc[x, 19] in hitIDs:
        updatedhitTF.append[1]
    else:
        updatedhitTF.append(0)

tracks['hitTF'] = updatedhitTF
tracks.head()
```

```python
# doing the outer join
# remember to use the proper spotify id col name
all_tracks_dataset = pd.merge(allbbs, tracks, on='spotifID', how='outer')
# the non-hit ones would be lacking values in 'date' and 'chart_position' cols
# for the former, the nonsense date 1/1/1900 can be used
# for the later, 0s can be used
all_tracks_dataset = all_tracks_dataset.fillna({'date': 1/1/1900, 'chart_position': 0})
all_tracks_dataset.head()
```
