```python
import pandas as pd
import numpy as np
```

```python
bb_tracks = pd.read_csv('../data/Hot Stuff.csv')
```

```python
bb_tracks.head()
```

```python
bb_tracks['WeekID'] = pd.to_datetime(bb_tracks['WeekID'])
```

```python
bb_tracks.dtypes
```

```python
# removing data from in or after 2018
bb_tracks_before_2018 = bb_tracks[bb_tracks['WeekID'].dt.date.astype(str) <= '2017-12-31']
bb_tracks_before_2018.head()
```

```python
bb_tracks_before_2018['WeekID'].unique()
```

```python
datetest = bb_tracks_before_2018.sort_values(['WeekID'])
```

```python
datetest.tail()
```

```python
bb_tracks_before_2018.columns
```

```python
bb_tracks_before_2018.columns = ['url', 'date', 'chart_position', 'song', 'performer', 'id',
       'Instance', 'previous_position', 'peak',
       'weeks_on_chart']
```

```python
bb_tracks_before_2018.head()
```

```python
bb_tracks_before_2018['hitTF'] = 1
```

```python
bb_tracks_before_2018.head()
```

```python
bb_tracks_before_2018 = bb_tracks_before_2018[['song', 'performer', 'chart_position', 'previous_position', 'peak', 'weeks_on_chart', 'hitTF', 'id']]                                               
```

```python
bb_tracks_before_2018.head()
```

```python
spotify = pd.read_csv('Hot 100 Audio Features.csv')
```

```python
spotify.head()
```

```python
spotify.columns
```

```python
spotify = spotify[['SongID', 'Performer', 'Song', 'spotify_track_id','spotify_track_duration_ms',
       'spotify_track_popularity', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']]
```

```python
spotify.head()
```

```python
spotify.isnull().sum()
```

```python
spotify.shape
```

```python
spotify.dropna(how='any', inplace=True)
```

```python
spotify.isnull().sum()
```

```python
# join where there is info either only in the spotify dataset (meaning has not appeared on billboard)
# or it is in both
joinedpre2018 = pd.merge(bb_tracks_before_2018,spotify,how='right', left_on=['id'],right_on=['SongID'])
```

```python
joinedpre2018.head()
```

```python
joinedpre2018['song'] = [x for x in joinedpre2018['Song']]
joinedpre2018['performer'] = [x for x in joinedpre2018['Performer']]
```

```python
joinedpre2018.head()
```

```python
joinedpre2018.isnull().sum()
```

```python
joinedpre2018 = joinedpre2018.fillna(0)
```

```python
joinedpre2018.head()
```

```python
joinedpre2018.columns
```

```python
joinedpre2018 = joinedpre2018[['song', 'performer', 'chart_position', 'previous_position', 'peak',
       'weeks_on_chart', 'hitTF',
       'spotify_track_id', 'spotify_track_duration_ms',
       'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']]
```

```python
joinedpre2018.head()
```

```python
joinedpre2018.columns = ['song', 'performer', 'chart_position', 'previous_position', 'peak',
       'weeks_on_chart', 'hitTF',
       'id', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms', 'time_signature']
```
